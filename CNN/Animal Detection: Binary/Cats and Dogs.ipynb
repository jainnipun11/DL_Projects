```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
train_dir=r'C:\Users\hp\Downloads\Data_set__Presentation_and_Python_Notebook_for_Image_Recognition_Project_Module\Data_set_Presentation_and_Python_Notebook_for_Image_recognition_project_module\cats_and_dogs\train'
valid_dir=r'C:\Users\hp\Downloads\Data_set__Presentation_and_Python_Notebook_for_Image_Recognition_Project_Module\Data_set_Presentation_and_Python_Notebook_for_Image_recognition_project_module\cats_and_dogs\validation'
test_dir=r'C:\Users\hp\Downloads\Data_set__Presentation_and_Python_Notebook_for_Image_Recognition_Project_Module\Data_set_Presentation_and_Python_Notebook_for_Image_recognition_project_module\cats_and_dogs\test'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen=ImageDataGenerator(
rescale=1./255,
height_shift_range=0.2,
width_shift_range=0.2,
rotation_range=40,
shear_range=0.2,
zoom_range=0.2,
horizontal_flip=True)

test_datagen=ImageDataGenerator(rescale=1./255)

train_generator=train_datagen.flow_from_directory(
train_dir,
target_size=(150,150),
batch_size=20,
class_mode='binary')

validation_generator=test_datagen.flow_from_directory(
valid_dir,
target_size=(150,150),
batch_size=20,
class_mode='binary')
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
from tensorflow.keras.applications import InceptionResNetV2
incept_conv=InceptionResNetV2(weights='imagenet',
                              include_top=False,
                             input_shape=(150,150,3))
```

```python
model=models.Sequential()
model.add(incept_conv)
model.add(layers.Flatten())
model.add(layers.Dense(256,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
from tensorflow.keras import optimizers
model.compile(loss='binary_crossentropy',
             optimizer=optimizers.RMSprop(learning_rate=2e-5),
             metrics=['acc'])
```

```python
from tensorflow import keras
checkpoint_cb=keras.callbacks.ModelCheckpoint("CNN2Project_Model-{epoch:02d}.h5")
```

```python
model_history=model.fit(
train_generator,
steps_per_epoch=30,
epochs=60,
validation_data=validation_generator,
validation_steps=30,
callbacks=[checkpoint_cb])
```
